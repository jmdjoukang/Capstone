```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Load the data
df = pd.read_csv('weather_data_2024.csv')
```

Next, we need to define the features and target variable. In this example, I want to predict 'Avg_Temperature_(°F)' based on the other features:

```python
# Define features and target
features = df.drop(['Date', 'Avg_Temperature_(°F)'], axis=1)
target = df['Avg_Temperature_(°F)']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

Now, we can create and train GradientBoostingRegressor model:

```python
# Create a GradientBoostingRegressor model
model = GradientBoostingRegressor(random_state=42)
```

```python
# Train the model
model.fit(X_train, y_train)
```

Finally, we can make predictions and evaluate the model:

```python
# Make predictions
predictions = model.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, predictions)
print("MSE: ", mse)
```
